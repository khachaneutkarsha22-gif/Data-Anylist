```python
#Feature Engineering - Age Groups & Tenure Buckets

bins = [18, 25, 35, 45, 55, 65]
labels = ['18-24','25-34','35-44','45-54','55+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
# Tenure bucket
bins_t = [0,2,5,10,40]
labels_t = ['0-1','2-4','5-9','10+']
df['TenureGroup'] = pd.cut(df['YearsAtCompany'], bins=bins_t, labels=labels_t, right=False)
```
